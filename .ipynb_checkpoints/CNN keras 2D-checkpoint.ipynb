# 2. Data loader

```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
class Data_loader():
    def __init__(self, data_train_dir, data_val_dir, data_test_dir):
        self.data_train_dir = data_train_dir
        self.data_val_dir = data_val_dir
        self.data_test_dir = data_test_dir
        
    def load_dataset(self):
        
        #Load dataset
        data_train = tf.keras.utils.image_dataset_from_directory(self.data_train_dir, batch_size=32, label_mode='binary', labels ="inferred", seed=42)
        data_val = tf.keras.utils.image_dataset_from_directory(self.data_val_dir, batch_size=32, label_mode='binary', labels ="inferred", seed=42)
        data_test = tf.keras.utils.image_dataset_from_directory(self.data_test_dir, batch_size=32, label_mode='binary', labels ="inferred", seed=42)
        
        #Visualize
        data_iterator_train = data_train.as_numpy_iterator()
        data_iterator_val = data_val.as_numpy_iterator()
        data_iterator_test = data_test.as_numpy_iterator()
        batch_train = data_iterator_train.next()
        batch_val = data_iterator_val.next()
        batch_test = data_iterator_test.next()
        print('Data train samples')
        fig, ax = plt.subplots(ncols=4, figsize=(20,20))
        for idx, img in enumerate(batch_train[0][:4]):
            ax[idx].imshow(img.astype(int))
            ax[idx].title.set_text(batch_test[1][idx]) 
        return data_train, data_val, data_test
    
    def fit(self):
        self.data_train, self.data_val, self.data_test = self.load_dataset()
        
    
```

```python
loader = Data_loader(data_train_dir=data_dir_list[0], data_val_dir=data_dir_list[1], data_test_dir=data_dir_list[2])
loader.fit()
```

```python
data_train, data_val, data_test = loader.load_dataset()
```

```python
import tensorflow as tf
data_train = tf.keras.utils.image_dataset_from_directory(data_dir_list[0], batch_size=32, label_mode='binary', labels ="inferred", seed=42)
data_val = tf.keras.utils.image_dataset_from_directory(data_dir_list[1], batch_size=32, label_mode='binary', labels ="inferred", seed=42)
data_test = tf.keras.utils.image_dataset_from_directory(data_dir_list[2], batch_size=32, label_mode='binary', labels ="inferred", seed=42)
```

```python
data_iterator_train = data_train.as_numpy_iterator()
data_iterator_val = data_val.as_numpy_iterator()
data_iterator_test = data_test.as_numpy_iterator()
```

```python
batch_train = data_iterator_train.next()
batch_val = data_iterator_val.next()
batch_test = data_iterator_test.next()
```

```python

fig, ax = plt.subplots(ncols=4, figsize=(20,20))
for idx, img in enumerate(batch_test[0][:4]):
    ax[idx].imshow(img.astype(int))
    ax[idx].title.set_text(batch_test[1][idx]) 
```

# 3. Rescale Images

```python
data_train = data_train.map(lambda x,y: (x/255, y))
data_train.as_numpy_iterator().next()
data_val = data_val.map(lambda x,y: (x/255, y))
data_val.as_numpy_iterator().next()
data_test = data_test.map(lambda x,y: (x/255, y))
data_test.as_numpy_iterator().next()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout, BatchNormalization
```

```python
model = Sequential()
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())
model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(1024, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
import tensorflow_addons as tfa
from tensorflow.keras.optimizers import Adam
import warnings
warnings.filterwarnings("ignore")
opt = Adam(learning_rate=0.0001, decay=0.02)
model.compile(optimizer=opt, loss=tf.losses.BinaryCrossentropy(), metrics=['accuracy', 
                                        tfa.metrics.FBetaScore(num_classes=2,average='micro',threshold=0.5),
                                        tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])
```

```python
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
pre = Precision()
re = Recall()
acc = BinaryAccuracy()
import warnings
warnings.filterwarnings("ignore")
for batch in data_test.as_numpy_iterator(): 
    X, y = batch
    yhat = model.predict(X)
    pre.update_state(y, yhat)
    re.update_state(y, yhat)
    acc.update_state(y, yhat)
print('>Precision:', pre.result())
print('>Recall:', re.result())
print('>F1 score:',  2*pre.result()*re.result()/(pre.result()+re.result()))
print('>BinaryAccuracy:', acc.result())
```
